```python
# Q 1 Answer 1:
"""
Both Probability Mass Function (PMF) and Probability Density Function (PDF) are ways of describing the probability distribution of a random variable.

The Probability Mass Function (PMF) is used to describe the probability distribution of a discrete random variable. 
It assigns probabilities to each possible value that the random variable can take. In other words, 
the PMF gives the probability of each outcome for a discrete random variable. The PMF is defined as:

PMF(x) = P(X=x)

Where X is a discrete random variable and x is a specific value that X can take.

An example of a PMF is the distribution of the outcomes of rolling a fair six-sided die. 
Since the possible outcomes of a die roll are 1, 2, 3, 4, 5, and 6, the PMF for this random variable would be:

PMF(1) = 1/6
PMF(2) = 1/6
PMF(3) = 1/6
PMF(4) = 1/6
PMF(5) = 1/6
PMF(6) = 1/6

The sum of all the probabilities in a PMF must equal 1.

On the other hand, Probability Density Function (PDF) is used to describe the probability distribution of a continuous random variable. 
The PDF does not assign probabilities to specific values of the random variable, 
but instead gives the probability density at each possible value. 
The area under the PDF curve between two values represents the probability that the random variable falls between those values. The PDF is defined as:

f(x) = dF(x) / dx

Where F(x) is the cumulative distribution function of the random variable X.

An example of a PDF is the normal distribution, which is often used to model continuous random variables in many real-world applications. 
The PDF for the normal distribution is:

f(x) = (1 / (σ * sqrt(2π))) * e^(-((x-μ)^2)/(2σ^2))

Where μ is the mean of the distribution, σ is the standard deviation, 
and e is the base of the natural logarithm.

The area under the normal distribution curve between two values represents the probability that the random variable falls between those values. 
The total area under the curve is equal to 1, meaning that the total probability of all possible outcomes is 1.

"""
```

```python
# Q 9 Answer 9:
"""
The central limit theorem (CLT) is a fundamental concept in statistics that states that the sampling distribution 
of the sample means or sample proportions of a random variable will approach a normal distribution as the sample size increases, 
regardless of the distribution of the underlying population.

In other words, if we take a large number of samples from any population, 
regardless of its distribution, and calculate the mean or proportion of each sample, 
the distribution of these sample means or sample proportions will be approximately normal, 
with a mean equal to the population mean and a standard deviation equal to the population standard deviation divided by the square root of the sample size.

The significance of the central limit theorem is that it allows us to use the normal distribution to make statistical inferences about population parameters, 
even if the underlying population distribution is not normal. This is because the normal distribution is well-understood and has many useful properties that make it easier to perform calculations and make predictions.

The central limit theorem has many practical applications in statistics and data analysis, including hypothesis testing, 
confidence interval estimation, and quality control. It also forms the basis for many advanced statistical techniques, 
such as regression analysis and analysis of variance.

Overall, the central limit theorem is a powerful and widely used concept in statistics that enables us to make accurate and reliable inferences about population parameters, 
even when we have limited information about the underlying population distribution.
"""
```
