```python
#Question 1 Answer:
"""
HTTP (Hypertext Transfer Protocol) is a protocol that defines how information is exchanged over the internet. 
There are many different methods defined in HTTP, but two of the most commonly used are the GET and POST methods.

GET Method:
The GET method is used to retrieve information from a server. When a client (such as a web browser) makes a GET request, 
it is asking the server to send back a representation of a resource (such as a web page). The request parameters are sent in the URL as a query string. The response from the server can be cached by the client, which can improve performance.

In simple terms, the GET method is used to retrieve data from the server. For example, when you visit a website, 
your browser sends a GET request to the server to fetch the web page.

POST Method:
The POST method is used to submit an entity (such as form data) to the server. When a client makes a POST request,
it is asking the server to accept the entity enclosed in the request (in the message body) and process it. 
The response from the server is typically a status code indicating whether the request was successful or not.

In simple terms, the POST method is used to send data to the server. For example, when you submit a form on a website, 
your browser sends a POST request to the server with the form data.

The key difference between GET and POST methods is that GET requests are used to retrieve data from the server,
while POST requests are used to send data to the server. Additionally, GET requests can be cached, while POST requests cannot be cached.

It's important to use the correct HTTP method for each request to ensure that the server can process the request correctly.
For example, you should use GET for requests that retrieve data and POST for requests that submit data to the server.

"""
```

```python
#Question 4 Answer:
"""
In Flask, templates are used to generate dynamic HTML pages that can be returned to the user. A template is a file that contains a mix of HTML and placeholders for dynamic content,
such as variables or expressions. When the template is rendered, 
the placeholders are replaced with the actual content to produce a complete HTML page.

The render_template() function is used in Flask to render a template and return the resulting HTML to the user. 
This function takes the name of the template file as its argument, along with any data that is needed to fill in the placeholders.

from flask import Flask, render_template

app = Flask(__name__)

@app.route('/')
def index():
    name = 'John'
    return render_template('index.html', name=name)


In this example, when a user visits the root URL of the application, the index() function is called. 
This function renders the index.html template, 
passing in the value of the name variable as a parameter. 
The index.html template might look something like this:

<!DOCTYPE html>
<html>
  <head>
    <title>Welcome to my site</title>
  </head>
  <body>
    <h1>Hello, {{ name }}!</h1>
    <p>Welcome to my site. We hope you enjoy your stay!</p>
  </body>
</html>



"""
```
