```python
# Q 1 Answer :
import scipy.stats as stats

sample_mean = 50
sample_std = 5
sample_size = 100 # assuming a sample size of 100

# calculate the standard error of the mean
std_error = sample_std / (sample_size**0.5)

# calculate the t-value for a 95% confidence interval with 99 degrees of freedom
t_value = stats.t.ppf(0.975, df=99)

# calculate the lower and upper bounds of the confidence interval
lower = sample_mean - t_value * std_error
upper = sample_mean + t_value * std_error

# print the results
print("95% Confidence Interval: [{:.2f}, {:.2f}]".format(lower, upper))
```

```python
# Q 2  Answer :
import numpy as np
from scipy.stats import chisquare

# observed frequencies of colors in a bag of M&Ms
observed = np.array([10, 15, 20, 5, 8, 12])

# expected frequencies of colors based on a distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown
expected = np.array([0.2, 0.2, 0.2, 0.1, 0.1, 0.2]) * np.sum(observed)

# perform the chi-square goodness of fit test
test_statistic, p_value = chisquare(observed, f_exp=expected)

# print the results
print("Test Statistic: {:.2f}".format(test_statistic))
print("p-value: {:.4f}".format(p_value))
```

```python
# Q 4  Answer :
import statsmodels.stats.proportion as proportion

# sample size and number of smokers
n = 500
x = 60

# calculate the 95% confidence interval for the true proportion of smokers
ci_low, ci_high = proportion.proportion_confint(x, n, alpha=0.05)

# print the results
print("95% Confidence Interval: [{:.4f}, {:.4f}]".format(ci_low, ci_high))
```

```python
# Q 5  Answer :

import scipy.stats as stats

# sample mean and standard deviation
mean = 75
std_dev = 12

# sample size
n = 1  # since we only have one sample, the sample size is 1

# calculate the 90% confidence interval for the population mean
ci_low, ci_high = stats.norm.interval(0.90, loc=mean, scale=std_dev / n**0.5)

# print the results
print("90% Confidence Interval: [{:.2f}, {:.2f}]".format(ci_low, ci_high))
```

```python
# Q 6  Answer :
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# set the degrees of freedom
df = 10

# create the chi-square distribution
x = np.linspace(stats.chi2.ppf(0.001, df), stats.chi2.ppf(0.999, df), 100)
y = stats.chi2.pdf(x, df)

# create the plot
fig, ax = plt.subplots()
ax.plot(x, y, 'b-', lw=2, label='chi-square pdf')

# shade the area corresponding to a chi-square statistic of 15
ax.fill_between(x[x>=15], y[x>=15], color='gray', alpha=0.5)

# set the x and y axis labels
ax.set_xlabel('chi-square statistic')
ax.set_ylabel('probability density')

# set the plot title
ax.set_title('Chi-Square Distribution with 10 Degrees of Freedom')

# display the plot
plt.show()


```

```python
# Q 7  Answer :

import statsmodels.api as sm
import numpy as np

# sample size
n = 1000

# number of people in the sample who prefer Coke
x = 520

# calculate the sample proportion
p_hat = x / n

# set the significance level and calculate the z-value
alpha = 0.01  # 99% confidence level
z_value = sm.stats.proportions_ztest(x, n, value=0)

# calculate the margin of error
margin_of_error = z_value[1] * np.sqrt(p_hat*(1-p_hat)/n)

# calculate the confidence interval
ci_low = p_hat - margin_of_error
ci_high = p_hat + margin_of_error

# print the results
print("99% Confidence Interval: [{:.4f}, {:.4f}]".format(ci_low, ci_high))
```

```python
# Q 8  Answer :


from scipy.stats import chisquare

# expected frequencies for a fair coin
expected_freq = [0.5, 0.5]

# observed frequencies
observed_freq = [55, 45]

# perform chi-square test
chi_sq, p_value = chisquare(observed_freq, expected_freq)

# set significance level
alpha = 0.05

# compare p-value to alpha
if p_value < alpha:
    print("Reject null hypothesis, the coin is biased.")
else:
    print("Fail to reject null hypothesis, the coin is fair.")

# print results
print("Chi-Square Statistic:", chi_sq)
print("P-Value:", p_value)
```

```python
# Q 9  Answer :
import numpy as np
from scipy.stats import chi2_contingency

# contingency table
observed_freq = np.array([[60, 140], [30, 170]])

# perform chi-square test for independence
chi_sq, p_value, dof, expected_freq = chi2_contingency(observed_freq)

# set significance level
alpha = 0.05

# compare p-value to alpha
if p_value < alpha:
    print("Reject null hypothesis, there is a significant association between smoking status and lung cancer diagnosis.")
else:
    print("Fail to reject null hypothesis, there is no significant association between smoking status and lung cancer diagnosis.")

# print results
print("Chi-Square Statistic:", chi_sq)
print("P-Value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected_freq)
```

```python
# Q 10  Answer :


import numpy as np
from scipy.stats import chi2_contingency

# contingency table
observed_freq = np.array([[200, 150, 150], [225, 175, 100]])

# perform chi-square test for independence
chi_sq, p_value, dof, expected_freq = chi2_contingency(observed_freq)

# set significance level
alpha = 0.01

# compare p-value to alpha
if p_value < alpha:
    print("Reject null hypothesis, there is a significant association between chocolate preference and country of origin.")
else:
    print("Fail to reject null hypothesis, there is no significant association between chocolate preference and country of origin.")

# print results
print("Chi-Square Statistic:", chi_sq)
print("P-Value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected_freq)
```
