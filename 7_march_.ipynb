```python
# Q 1 Answer 1:
"""
The three measures of central tendency are:

1)Mean: The mean is the average of a set of numbers. 
It is calculated by summing up all the values in the dataset and dividing by the total number of values.

2)Median: The median is the middle value in a sorted dataset. 
It is the value that separates the dataset into two halves, 
with half of the values being less than the median and half of the values being greater than the median.

3)Mode: The mode is the value that appears most frequently in a dataset. 
In other words, it is the value that has the highest frequency or count in the dataset.

These measures are used to describe the central tendency or typical value of a dataset. 
They are all useful in different ways and are often used together to get a more complete picture of the data. 
The choice of which measure to use depends on the nature of the data and the research question being addressed.

"""
```

```python
# Q 8 Answer 8:
"""
Skewness in data refers to the degree of asymmetry of a probability distribution.
A distribution is said to be symmetric if it is roughly the same on both sides of the center (mean, median, or mode). 
A positively skewed distribution has a long tail on the right side, and the majority of the data values are concentrated on the left side.
In contrast, a negatively skewed distribution has a long tail on the left side, 
and the majority of the data values are concentrated on the right side.

Skewness is an important aspect of data analysis as it can impact the validity of statistical tests and the interpretation of results. 
In general, when data is skewed, it may not be appropriate to use the mean as a measure of central tendency, 
as it can be influenced by extreme values in the tail. Instead, the median may be a better measure of central tendency for skewed data.

Skewness can be measured using various statistical methods, such as the skewness coefficient or skewness test. 
Understanding the degree and direction of 
skewness in a dataset can help researchers make more accurate conclusions and predictions based on the data.
"""
```

```python
# Q 9 Answer 9:
"""
If a dataset is right skewed, meaning it has a long tail on the right side, the median will be less than the mean. 
This is because the mean is affected by the extreme values in the tail, pulling it towards the right side and increasing its value.
In contrast, the median represents the middle value of the dataset and is less affected by extreme values. 
As a result, it will be closer to the values that are concentrated on the left side of the distribution,
which will be relatively smaller than the values in the long right tail.
"""
```

```python
# Q 10 Answer 10:
"""
Covariance and correlation are both measures of the relationship between two variables in a dataset. However, 
they differ in how they express this relationship.

Covariance measures the degree to which two variables vary together. It is a measure of the joint variability of two random variables. 
If the covariance between two variables is positive, then they tend to increase or decrease together. 
If the covariance is negative, then as one variable increases, the other variable decreases. 
A covariance of zero indicates that the two variables are independent, meaning they have no linear relationship.

On the other hand, correlation measures the strength and direction of the linear relationship between two variables. 
It is a standardized version of covariance, and it ranges from -1 to 1. 
A correlation of -1 indicates a perfect negative linear relationship,
a correlation of 0 indicates no linear relationship, and a correlation of 1 indicates a perfect positive linear relationship.

In statistical analysis, covariance and correlation are used to identify the strength and direction of the relationship between two variables.
They can help determine whether a relationship is positive or negative, strong or weak. These measures can also help identify outliers,
which can have a significant impact on the covariance and correlation values. 
Covariance and correlation are often used in regression analysis to predict the value of one variable based on the value of another variable.

Overall, covariance and correlation are both useful measures in statistical analysis, 
but correlation is often preferred because it provides a standardized measure of the linear relationship between two variables, 
which makes it easier to compare across different datasets.

"""
```
