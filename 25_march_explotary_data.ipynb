```python
# Q3  Answer :

# Get the flight prices
prices = df['Price']

# Find the minimum and maximum prices
min_price = prices.min()
max_price = prices.max()

# Compute the range of prices
price_range = max_price - min_price

# Print the results
print("Minimum price:", min_price)
print("Maximum price:", max_price)
print("Price range:", price_range)
```

```python
# Q 13 Answer:
# Filter the dataset to include only the necessary columns
apps_type_price = df[['Type', 'Price']].dropna()

# Convert the price values to a numeric format
prices = []
for price in apps_type_price['Price']:
    try:
        prices.append(float(price[1:]))
    except ValueError:
        prices.append(0)
apps_type_price['Price'] = prices

# Group the data by app type and calculate the mean price
apps_type_price_mean = apps_type_price.groupby('Type').mean()

# Create the bar chart using Seaborn
sns.barplot(x=apps_type_price_mean.index, y='Price', data=apps_type_price_mean)
```

```python
# Q 14 Answer:
category = pd.DataFrame(df['Category'].value_counts())        #Dataframe of apps on the basis of category
category.rename(columns = {'Category':'Count'},inplace=True)
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x=category.index[:10], y ='Count',data = category[:10],palette='hls')
plt.title('Top 10 App categories')
plt.xticks(rotation=90)
plt.show()
```
